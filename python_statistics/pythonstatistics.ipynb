```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Numpy Fun

```python
randomarray = np.random.randint(1000, size=100)

print(randomarray)
print(len(randomarray))
print(type(randomarray))

plt.style.use('ggplot')
```

```python
print("Mean" , sum(randomarray)/len(randomarray))
print("Mean" , np.mean(randomarray))
print("SD" , np.std(randomarray))
print("Min" , np.min(randomarray))
print("Max" , np.max(randomarray))
print("25%" , np.percentile(randomarray,25))
print("75%" , np.percentile(randomarray,75))
```

```python
plt.hist(randomarray, edgecolor="blue", bins=50)
plt.title("RandomArray1 Histogram (100 samples between 1-1,000)")
plt.show()
```

```python
randomarray2 = np.random.randint(1000, size=1000000)
print(randomarray2)
print(len(randomarray2))
print(type(randomarray2))
```

```python
print("Mean" , sum(randomarray2)/len(randomarray2))
print("Mean" , np.mean(randomarray2))
print("SD" , np.std(randomarray2))
print("Min" , np.min(randomarray2))
print("Max" , np.max(randomarray2))
print("25%" , np.percentile(randomarray2,25))
print("75%" , np.percentile(randomarray2,75))
```

```python
plt.hist(randomarray2, edgecolor="blue", bins=50)
plt.title("RandomArray2 Histogram (1,000,000 samples between 1-1,000)")
plt.show()
```

```python
# RandomArray2 has more even distibution becaus it has 1,000,000 samples. This makes the results more consistent. 
```

## Data Exploration/Analysis

```python
data = pd.read_csv("2021-PM2.5.csv", parse_dates=['Date'])
data
```

```python
data.describe()
```

```python
type(data["HW-MC"])
```

```python
saltlakeairdata = data["HW-MC"]
```

```python
print(len(saltlakeairdata))
print(np.min(saltlakeairdata))
print(np.max(saltlakeairdata))
print(np.mean(saltlakeairdata))
print(np.median(saltlakeairdata))
```

```python
data[::20].plot.scatter(x=0, y=1, ylabel="saltlakeairdata",
                        title="Salt Lake City Air Particulate Data 2021")
```

```python
data["HW-MC"].hist(bins=(30))
```

```python
data["HW-MC"].hist()
```

```python
monthlydata = data["HV-MC"].groupby([data["Date"].dt.month]).mean()
monthlydata
```

```python
monthlydata.plot()
```

```python
hourlydata = data["HV-MC"].groupby([data["Date"].dt.hour]).mean()
hourlydata
```

```python
hourlydata.plot()
```
